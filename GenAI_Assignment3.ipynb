```python
# Step 1: Install Required Libraries
!pip install langchain openai langchain-openai langchain-core langchain-community langchain-classic

# Step 2: Import Required Modules
from langchain_openai import OpenAI
from langchain_core.prompts import PromptTemplate
from langchain_classic.chains import LLMChain, SequentialChain
```

```python
# Step 3: Initialize the Language Model
from getpass import getpass

OPENAI_API_KEY = getpass("Enter your OpenAI API key:")
llm = OpenAI(openai_api_key=OPENAI_API_KEY, temperature=0.3)
```

```python
# Step 4: Create Prompt Template for Reformulation
rephrase_prompt = PromptTemplate(
    input_variables=["text"],
    template="Rewrite the following text in a clear and professional way:\n{text}"
)
```

```python
# Step 5: Create Prompt Template for Summarization
summary_prompt = PromptTemplate(
    input_variables=["rephrased_text"],
    template="Provide a concise summary of the following text:\n{rephrased_text}"
)
```

```python
# Step 6: Create Individual Chains
rephrase_chain = LLMChain(
    llm=llm,
    prompt=rephrase_prompt,
    output_key="rephrased_text"
)

summary_chain = LLMChain(
    llm=llm,
    prompt=summary_prompt,
    output_key="summary"
)
```

```python
# Step 7: Combine Chains into a Sequential Pipeline
overall_chain = SequentialChain(
    chains=[rephrase_chain, summary_chain],
    input_variables=["text"],
    output_variables=["rephrased_text", "summary"]
)
```

```python
# Step 8: Run the Pipeline
input_text = "Artificial intelligence is transforming industries by automating processes and improving decision-making."

result = overall_chain({"text": input_text})

print("Rephrased Text:\n", result["rephrased_text"])
print("\nSummary:\n", result["summary"])
```
